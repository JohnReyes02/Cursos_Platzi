# Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['INDIA', 'JAPAN', 'MEXICO', 'COLOMBIA', 'GERMANY']
population = [1000, 800, 900, 1000, 300]
```

Barras verticales

```python
plt.bar(country, population, width=0.5, color=['red', 'blue']) # width para ancho de columnas
plt.xticks(np.arange(5), ('India', 'Japon', 'Mexico', 'Colombia', 'Alemania'), rotation=45)
plt.show()
```

Barras Horizontales

```python
plt.barh(country, population)
plt.show()
```

Reto

```python
plt.bar(country, population, width=0.5, color='cyan') # width para ancho de columnas
plt.xticks(np.arange(5), ('India', 'Japon', 'Mexico', 'Colombia', 'Alemania'))
plt.title('Grfico de barras de poblacin')
plt.xlabel('Pases')
plt.ylabel('Poblacin')
plt.show()
```
