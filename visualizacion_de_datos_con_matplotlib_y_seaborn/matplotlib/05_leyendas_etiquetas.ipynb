# Leyendas, etiquetas, títulos, tamaño

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(5,5))

axes[0].plot(x, y, label='$sin(x)$')
axes[0].set_title('Relacion X - Y') # Para colocar un titulo a la figura
axes[0].set_xlabel('X') # Para colocar el nombre al eje x
axes[0].set_ylabel('Y') # Para cololar el nombre al eje y
axes[0].legend() # para mostrar los labels definidos arriba

axes[1].plot(y, x)
axes[1].set_title('Relacion Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

Sin el método orientado a objetos

```python
plt.plot(x, y, label='$sin(x)$')
plt.title('Este es un titulo')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='lower left', bbox_to_anchor=(1,0.9))
plt.show()
```
