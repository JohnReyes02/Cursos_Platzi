Un objeto define una figura, esa figura es un lienzo en el cual podemos introducri diferenctes grficas(axes), de las cuales cada una posee sus propios ejes (axis).

La figura representa el todo, dentro de ella vamos a configurara grficas las cuales conteine diferentes ejes.



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

### Mtodo orientado a objetos

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.5, 0.9])
axes.plot(x, y, 'b')
fig.show()
```

Esta forma me da ventajas cuando quiero hacer algo ms especfico.

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.9]) # [ubicacion ubicacion tamao tamao]
axes2 = fig.add_axes([0.2, 0.55, 0.4, 0.3])

axes.plot(x, y, 'b')
axes2.plot(y, x, 'r')
fig.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.9])
axes2 = fig.add_axes([0.2, 0.55, 0.4, 0.3])

axes.plot(x, y, 'b')
axes2.plot(y, x, 'r')
axes2.set_facecolor('gray')
fig.show()
```
