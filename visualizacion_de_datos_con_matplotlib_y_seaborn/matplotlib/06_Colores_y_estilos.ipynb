## Colores y Estilos

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
print(plt.style.available) # Para ver los diferentes estilos que podemos dar a las gráficas
```

```python
#plt.style.use('bmh') # seleccionamos un estilo para la gráfica
plt.style.use('dark_background') # 
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

ax.plot(x, x+1, 'r--')
ax.plot(x, x+2, 'b-')
ax.plot(x, x+3, 'y.-')
ax.plot(x, x+4, 'go:')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

ax.plot(x, x+1, color='green', alpha=0.8, linewidth=5, linestyle='-', marker='o', markersize=10,
        markerfacecolor='#CCFFCC')
ax.plot(x, x+2, color='blue', linewidth=8, linestyle='--', marker='x')
ax.plot(x, x+3, color='#66FF66', linewidth=3, linestyle='dashed', marker='v')
ax.plot(x, x+4, color='#CCFFCC', linewidth=2, linestyle=':', marker='p')
plt.show()
```
