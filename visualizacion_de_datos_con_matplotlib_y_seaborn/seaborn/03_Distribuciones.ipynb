# Distribuiciones

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## Gráfico de Histograma

```python
sns.histplot(data=tips, x='tip')
plt.show()
```

Se puede seleccionar la agrupacion de las barras con `bins`:

```python
sns.histplot(data=tips, x='tip', bins=15)
plt.show()
```

Se puede graficar de forma acumulativa:

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=True)
plt.show()
```

Clasificación por variable categórica:

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex')
plt.show()
```

Se puede ajustar con estadisticas como la frecuencia en lugar del conteo:

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='frequency')
plt.show()
```

Estadística de probabilidad:

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='probability')
plt.show()
```

Estadísitica por porcentaje:

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='percent')
plt.show()
```

Utilizamos el parámetro `multiple`:

- `stack`

- `layer`

- `dodge`

- `fill`

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='count', 
             multiple='dodge')
plt.show()
```

## Diagrama de Densidad

```python
sns.kdeplot(data=tips, x='tip', hue='sex')
plt.show()
```

Se puede ver el parámetro `cumulative`:

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=True)
plt.show()
```

Para dibujar el área bajo la curva usamos el parámetro `fill`:

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, 
            fill=True) # shade=True deprecated
plt.show()
```

Se puede usar el parámetro `bw_adjust` para ajustar

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, 
            fill=True, bw_adjust=1) 
plt.show()
```

## Diagrama de Acumulación

```python
sns.ecdfplot(data=tips, x='tip', hue='sex')
plt.show()
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex', stat='count')
plt.show()
```

## Diagramas de distribución

```python
sns.displot(data=tips, x='tip', hue='sex', kind='hist', multiple='stack')
plt.show()
```
